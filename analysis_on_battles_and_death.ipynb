```python
#Creating a relation between battles and locations
import pandas as pd

battles_df = pd.read_csv("battles.csv")
locations_df = battles_df[['location']].drop_duplicates().reset_index(drop=True)

# Creating a new dataframe to represent the relation between battles and locations
battle_location_df = battles_df[['name', 'location']].rename(columns={'name': 'battle_name'})

# Merging the location data into the relation table
battle_location_df = pd.merge(battle_location_df, locations_df, on='location', how='left')
battle_location_df.head()
```

```python
import matplotlib.pyplot as plt

battle_counts = battle_location_df.groupby(['location']).count()
battle_counts.plot.pie(y='battle_name', legend=False, autopct='%1.f%%', figsize=(8,8))
plt.ylabel('')
plt.title('Battles by Location')
plt.show()
```
